```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 1001)
y = np.cos(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-1, 1, 1001)
fig, axes = plt.subplots(2, 2)

for i, axis in enumerate(axes.flatten()):
    y1 = x ** i
    y2 = (x - 1) ** i
    
    axis.plot(x, y1)
    axis.plot(x, y2)
    
plt.show()
```
