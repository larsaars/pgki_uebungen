```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 1001)
fig, axes = plt.subplots(3, 2)

for n_min_1, axis in enumerate(axes.flatten()):
    n = n_min_1 + 1
    
    y1 = x**n
    y2 = n * x**n_min_1
    
    axis.plot(x, y1)
    axis.plot(x, y2)

plt.show()
```
